```python
#TypedDict
#A way to define dictionary in python where we define what keys and values should exist

from typing import TypedDict

class Person(TypedDict):
    name:str
    age:int

new:Person={'name':'Roshan','age':30}

new
```

```python
#pydantic
#it is a data validation and data parsing library

from pydantic import BaseModel

class Student(BaseModel):
    name: str

new_student={'name':'Roshan'}

student=Student(**new_student)
print(student)
```
